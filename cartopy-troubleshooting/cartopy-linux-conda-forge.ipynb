# Cartopy Orthographic Projection - Linux

```python
import pandas
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature

%matplotlib inline
```

```python
cities = pandas.read_csv('https://raw.githubusercontent.com/jenfly/eoas-python/master/data/country_capitals.csv')
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1, projection=ccrs.Orthographic())
ax.stock_img()
ax.add_feature(cfeature.COASTLINE)
ax.scatter(data=cities, x='Longitude', y='Latitude', marker='o', 
           color='magenta', transform=ccrs.PlateCarree());
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1, projection=ccrs.Orthographic())
ax.add_feature(cfeature.LAND)
ax.add_feature(cfeature.OCEAN);
```
